# Estimating amplitude and phase of a tone

```python
rng = np.random.default_rng()
```

```python
fs = 2
fc = 0.5
theta0 = rng.uniform(low=-np.pi, high=np.pi)
A = rng.uniform(low=math.sqrt(2*50*1e-3)/10, high=math.sqrt(2*50*1e-3)*2)

x = tonegen.tonegen(fs, fc, cossin="cos", theta0=theta0, nsamp=2048)
x = A*x

awgn = rf_analog.AWGN(-20, 1, 2)
y = awgn.transform(x)

# fig, ax = plt.subplots(dpi=100)
# ax.plot(x[:100])
# ax.plot(y[:100])
# ax.grid()
```

```python
A_hat, phi_hat = rf_est.est_tone_amp_phase(y, fs, fc)

print(A)
print(A_hat)
print(theta0)
print(phi_hat)
```
