## Continuous-time flip and shift

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
```

```python
t = np.arange(-5, 10, 1)
x = np.zeros_like(t)
x[(-2 <= t) & (t <= 5)] = 8-t[(-2 <= t) & (t <= 5)]
```

```python
colors = rcParams["axes.prop_cycle"].by_key()["color"]

shift = 2

fig, ax = plt.subplots(dpi=150)
ta = -2
tb = 5
# ax.plot(t, x, label="x(n)")
# ax.plot(-t, x, label="x(-n)")
# ax.plot(-t+shift, x, label=f"(x({shift}-n))")
ax.stem(t, x, label="x(n)", linefmt=colors[0])
ax.stem(-t, x, label="x(-n)", linefmt=colors[1])
ax.stem(-t+shift, x, label=f"(x({shift}-n))", linefmt=colors[2])


ax.grid()
ax.legend()
ax.set_xlim(left=-6, right=6)
ax.set_xticks(range(-6, 6))
ax.set_ylabel("y(t)")
ax.set_xlabel("t")
ax.vlines([ta, tb], ymin=0, ymax=15, colors=colors[0], linestyles="dotted", linewidth=2)
ax.vlines([shift-ta, shift-tb], ymin=0, ymax=15, colors=colors[2], linestyles="dotted", linewidth=2)
ax.set_title("Signal Transformation: Shift + Reverse")
```
