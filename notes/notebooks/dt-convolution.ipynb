# Signals and Systems for Dummies, Chapter 6

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
```

```python
t = np.arange(-1, 1.5, 1).astype("int")
x = np.array([1, 2, 3])
h = np.array([0, 2, 1])
c = signal.convolve(x, h, method="direct")
# t_c = np.arange(-2, -2+len(c), 1).astype("int") # first non-zero value is -(1 + t_x + t_h) where t_x and t_h are the starting indices for x and h (indices of first non-zero values)
t_c = np.arange(-2, 2+0.5, 1).astype("int")
```

```python
colors = rcParams["axes.prop_cycle"].by_key()["color"]
fig, ax = plt.subplots(dpi=150)
markerline, stemlines, baseline = ax.stem(t, x, label="x", linefmt=colors[0], )
plt.setp(markerline, markersize=15)
ax.stem(t, h, label="h", linefmt=colors[1])
ax.stem(t_c, c, label="x * h", linefmt=colors[2])
ax.grid()
ax.legend()
ax.set_xticks(t_c)
# ax.set_xlim(right=10)
ax.set_ylabel("y(n)")
ax.set_xlabel("n")
ax.set_title("Finite Extent Signal - Convolution Example")
```

## Convolving two rectangular pulses

```python
N = 10
t = np.arange(0, N, 1).astype("int")
x = np.ones_like(t)
h = np.ones_like(t)
c = signal.convolve(x, h, method="direct")
# t_c = np.arange(-2, -2+len(c), 1).astype("int") # first non-zero value is -(1 + t_x + t_h) where t_x and t_h are the starting indices for x and h (indices of first non-zero values)
t_c = np.arange(0, 2*(N-1)+0.5, 1).astype("int")
```

```python
colors = rcParams["axes.prop_cycle"].by_key()["color"]
fig, ax = plt.subplots(dpi=150)
# markerline, stemlines, baseline = ax.stem(t, x, label="x", linefmt=colors[0], )
# plt.setp(markerline, markersize=15)
# ax.stem(t, h, label="h", linefmt=colors[1])
ax.stem(t_c, c, label="x * h", linefmt=colors[2])
ax.grid()
ax.legend()
ax.set_xticks(t_c)
# ax.set_xlim(right=10)
ax.set_ylabel("y(n)")
ax.set_xlabel("n")
ax.set_title("Finite Extent Signal - Convolution Example")
```

## Convolving step with decaying exponential

```python
a = 0.8

t = np.arange(0, 20, 1)
x = np.ones_like(t)
h = a**t
y = (1-a**(t+1))/(1-a)
```

```python
colors = rcParams["axes.prop_cycle"].by_key()["color"]
fig, axs = plt.subplots(nrows=3, sharex=True, dpi=150, figsize=(3.2, 4.8))
axs[0].stem(t, x)
axs[0].grid()
axs[0].set_ylabel("x(n)")

axs[1].stem(t, h)
axs[1].grid()
axs[1].set_ylabel("h(n)")

axs[2].stem(t, y)
axs[2].grid()
# axs[2].set_xticks(t)
# ax.set_xlim(right=10)
axs[2].set_ylabel("y(n)")
axs[2].set_xlabel("n")
fig.suptitle("Input Step * Decaying Exponential")
```

## Convolving pulse with decaying exponential

```python
a = 0.8
N = 4

t = np.arange(0, 20, 1)
x = np.zeros_like(t)
x[:N] = 1
h = a**t
y = (1-a**(t+1))/(1-a)
y = np.zeros_like(t, dtype="float")
y[:N] = (1-a**(t[:N]+1))/(1-a)
y[N:] = a**(t[N:]-N+1)*(1-a**N)/(1-a)
```

```python
colors = rcParams["axes.prop_cycle"].by_key()["color"]
fig, axs = plt.subplots(nrows=3, sharex=True, dpi=150, figsize=(3.2, 4.8))
axs[0].stem(t, x)
axs[0].grid()
axs[0].set_ylabel("x(n)")

axs[1].stem(t, h)
axs[1].grid()
axs[1].set_ylabel("h(n)")

axs[2].stem(t, y)
axs[2].grid()
# axs[2].set_xticks(t)
# ax.set_xlim(right=10)
axs[2].set_ylabel("y(n)")
axs[2].set_xlabel("n")
fig.suptitle("Input Step * Decaying Exponential")
```
